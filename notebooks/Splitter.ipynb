## Split data

```python
import shutil
from pathlib import Path
from sklearn.model_selection import train_test_split

IMAGE_GLOB = "Potsdam_dataset/2_Ortho_RGB"
MASK_GLOB = "Potsdam_dataset/5_Labels_all"


def split_images_and_masks(image_glob, mask_glob, test_size=0.2, random_state=42):
    image_paths = list(Path(image_glob).glob("*.tif"))
    mask_paths = list(Path(mask_glob).glob("*.tif"))

    train_images, val_images, train_masks, val_masks = train_test_split(
        image_paths, mask_paths, test_size=test_size, random_state=random_state
    )
    return train_images, val_images, train_masks, val_masks


train_images, val_images, train_masks, val_masks = split_images_and_masks(
    IMAGE_GLOB, MASK_GLOB, test_size=0.2, random_state=42
)
```

```python
TRAIN_FOLDER = "Potsdam_dataset/train"
VAL_FOLDER = "Potsdam_dataset/val"


def create_split_folders(folder_name, images, masks):
    Path(folder_name).mkdir(exist_ok=True)

    for img, msk in zip(images, masks):
        destination_img = Path(folder_name) / "images" / img.name
        destination_msk = Path(folder_name) / "masks" / msk.name
        shutil.move(img, destination_img)
        shutil.move(msk, destination_msk)
```

```python
create_split_folders(TRAIN_FOLDER, images=train_images, masks=train_masks)
```

```python
create_split_folders(VAL_FOLDER, images=val_images, masks=val_masks)
```
